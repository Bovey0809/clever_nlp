```python
def filter_words_by_norm(norm_range=(0, 1.2)):
    data = pd.read_csv("wordnet_bert_common_words.csv")
    words = data[data.norms.apply(
        lambda x: norm_range[0] < x < norm_range[1])].words.values
    words = set(words.tolist())
    print(f">>>> There are {len(words)} words between norm range {norm_range}")
    return words
```

```python
words = filter_words_by_norm((1.6, 2))
data[data.word.apply(lambda x: x in words)].to_csv('low_freq.csv')
```

```python
high_freq_data = data[data.word.apply(lambda x: x in words)].drop(columns=['Unnamed: 0'])
```

- [ ] 取出来50个词，精简recall，让 recall 尽可能只含有单词。
- [x] 去掉所有 unusedxxx 
- [x] 塞选出所有的名词

```python
wordnet_nouns = list(wn.all_lemma_names(pos='n', lang='eng'))
```

```python
all_nouns = high_freq_data[high_freq_data.word.apply(lambda x: x in wordnet_nouns)]
```

```python
excluded_unusesed = all_nouns[all_nouns.norm_recall.apply(lambda x: 'unused' not in x)]
```

- [ ] 对 word 进行中文翻译

```python
excluded_unusesed['translate'] = excluded_unusesed.word.apply(lambda x: wn.synsets(x)[0].lemma_names('cmn'))
```

```python
def filter_numbers(x):
    x = x.replace('\'', '').replace('[', '').replace(']', '').replace(' ','')
    word_count = 0
    words = []
    for word in x.split(','):
        if not word.isdigit():
            word_count += 1
            words.append(word)
    return word_count > 5
filter_numbers_df = excluded_unusesed[excluded_unusesed.norm_recall.apply(filter_numbers)]
```

```python
filter_numbers_df
```

```python
reorder_df = filter_numbers_df[['word', 'translate', 'definition', 'norm_recall', 'original_recall']]
```

reorder_df = reorder_df.rename(columns={"word": "单词", "translate": "翻译", "definition": "字典释义", "norm_recall": "新召回", "original_recall": "原始召回"})

```python
reorder_df.norm_recall = reorder_df.norm_recall.apply(lambda x: x.split(',')[:5])
```

```python
reorder_df.original_recall = reorder_df.original_recall.apply(lambda x: x.split(',')[:5])
```

```python
def clean_format(x):
    return ''.join([i.strip('[]') for i in x])
reorder_df.original_recall = reorder_df.original_recall.apply(clean_format)
reorder_df.norm_recall = reorder_df.norm_recall.apply(clean_format)
```

```python
reorder_df.to_csv('final.csv')
```
